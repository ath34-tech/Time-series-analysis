```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly as pyp
import yfinance as yf
%matplotlib inline
```

```python
def get_data(ticker):
    data=yf.download(ticker,period="5y",interval="1d")
    return data
```

```python
df_hmc=get_data('HMC')
```

```python
df_hmc_date=df_hmc.reset_index()
```

```python
df_hmc_date.head()
```

```python
fig=px.line(df_hmc_date,x="Date",y="Open")
fig.update_layout(width=1000,height=600,paper_bgcolor="#fafafa",plot_bgcolor="white")
fig.update_xaxes(showgrid=True,gridwidth=0.5)
fig.update_xaxes(showgrid=True,gridwidth=0.5)

fig.show()
```

- we can see the data isn't stationary
- its have a lot of fluctuation
- we gotta do some more test and make it stationary for ARIMA forecasting

### Augmented Dicky-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(col_name):
    results=adfuller(df_hmc_date[col_name].dropna())
    ADF_stats=results[0]
    p_value=results[1]    
    print(f"ADF statistics:{ADF_stats}")
    print(f'p-value:{p_value}')
```

```python
adf_test("Open")
```

- The above p-value suggest data is non-stationary
- We will use differencing to analyze it more and make it stationary

### Differencing

```python
differenced_data_1 = df_hmc_date["Open"].diff().dropna()
```

```python
result_diff = adfuller(differenced_data_1)
```

```python
print('ADF Statistic (Differenced Data):', result_diff[0])
print('p-value (Differenced Data):', result_diff[1])
for key, value in result_diff[4].items():
    print('Critical Values (Differenced Data):')
    print(f'   {key}: {value}')
```

### Seasonality decompose

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(df_hmc["Open"], model='additive', period=252)  # 252 for daily data considering trading days in a year

# Plot the decomposed components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8))
decomposition.observed.plot(ax=ax1, title='Observed')
decomposition.trend.plot(ax=ax2, title='Trend')
decomposition.seasonal.plot(ax=ax3, title='Seasonal')
decomposition.resid.plot(ax=ax4, title='Residual')

plt.tight_layout()
plt.show()
```

### Forecasting

```python
#splitting training and test data
train_size = int(len(df_hmc_date["Open"]) * 0.8)
train_data, test_data = df_hmc_date["Open"][:train_size], df_hmc_date[train_size:]
```

```python
import pmdarima as pm
```

```python
auto_arima_model = pm.auto_arima(
    train_data, 
                          test='adf',       # use adftest to find optimal 'd'

    seasonal=False,  # Set to True if your data has a seasonal component
    m=1,  # The period for seasonal differencing, if applicable
    stepwise=True,  # Use a stepwise algorithm to search for the best model
    suppress_warnings=True,  # Suppress warnings for better readability
    trace=True,  # Set to True to print the progress
                          max_p=3, max_q=3, # maximum p and q

)
```

```python
print(auto_arima_model.summary())
```

```python
auto_arima_model.plot_diagnostics(figsize=(15,8))
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train_data, order=(2,1,2))  # Replace (p, d, q) with your values
model_fit = model.fit()
```

```python
print(model_fit.summary())
```

```python
forecast_result = model_fit.get_forecast(steps=321)
```

```python
fc = forecast_result.predicted_mean
conf = forecast_result.conf_int(alpha=0.05)  # 95% confidence interval
```

```python
fc_series = pd.Series(fc, index=test_data.index[:321])
lower_series = pd.Series(conf.iloc[:, 0], index=test_data.index[:321])
upper_series = pd.Series(conf.iloc[:, 1], index=test_data.index[:321])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(train_data, label='Training Data')
plt.plot(test_data["Open"], color='blue', label='Actual Stock Price')
plt.plot(fc_series, color='orange', label='Predicted Stock Price')
plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.10)
plt.title(f'Reliance Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel(f'Reliance Stock Price')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

### Model evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
fc_series = pd.Series(fc, index=test_data.index[:321])

# Ensure that the forecast and test data align
test_data = test_data[:321]["Open"]

# Calculate evaluation metrics
mae = mean_absolute_error(test_data, fc_series)
mse = mean_squared_error(test_data, fc_series)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
